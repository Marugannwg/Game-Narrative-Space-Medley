### Grouping continuous dialouges
In the gameplay, it is impossible to show a *long* dialogue at the same time. We need to combine multiple lines into one utterance when then are continuously by same character.

Concatenating continuous lines spoken by the same characters into single dialogue.

```python
dialogue_df['group'] = ((dialogue_df['Story'] != dialogue_df['Story'].shift()) | (dialogue_df['Character'] != dialogue_df['Character'].shift())).cumsum()

dialogue_df.head(40)
```

```python
dialogue_df
```

```python
grouped_df = dialogue_df.groupby(['Story', 'Character', 'group'], as_index=False)['Dialogue'].agg(' '.join) ## Same group is one utterance
```

```python
# order by group
grouped_df = grouped_df.sort_values(by=['group'])

grouped_df.head(40)
```

```python
grouped_df.reset_index(drop=True, inplace=True)
```

```python
grouped_df.head(40)
```

```python
# drop Option_1, Option_2, Option_3 among the Character column -- these are not characters 

grouped_df = grouped_df[~grouped_df['Character'].str.contains('Option_')]

grouped_df.head(40)
```

```python
# Drop all the ".xlsx" at the end of story name

grouped_df['Story'] = grouped_df['Story'].str.replace('.xlsx', '')

grouped_df.head(40)
```

```python
# see unique story names 
# NOTE the actual name is always after the last "_"

grouped_df['Story'].unique()
```

```python
# drop the group column
grouped_df = grouped_df.drop(columns=['group'])
```

```python
grouped_df.head(14)
```

```python
# save the grouped_df to a new csv file
grouped_df.to_csv('data\\Arknights_plot\\grouped_dialogues.csv', index=False)
```

```python
# count the number of dialogues for each character, sorted and top 20

grouped_df['Character'].value_counts().sort_values(ascending=False).head(40)
```

```python
# count how many words in total

grouped_df['word_count'] = grouped_df['Dialogue'].apply(lambda x: len(x.split()))

grouped_df['word_count'].sum()
```

```python
grouped_df.head(10)
```

```python
# rank characters by word_count.sum()

grouped_df.groupby('Character')['word_count'].sum().sort_values(ascending=False).head(40)
```

```python
# pick out the top 40 characters
top_40_characters = grouped_df.groupby('Character')['word_count'].sum().sort_values(ascending=False).head(40).index
# 'non-character' and '???' are not characters

top_40_characters = top_40_characters[~top_40_characters.isin(['non-character', '???'])]

top_40_characters
```
